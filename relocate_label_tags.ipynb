```python
import cv2
import numpy as np
import os
from tqdm import tqdm
```

```python
targets = {
    0: 0,  #taşit
    1: 1,  #insan
    2: 2,  #uap
    3: 3,  #uai
    4: 3
}
```

```python
def path_sequence(root,*directories):
    seq = root
    for directory in directories:
        seq += f"/{directory}"
    return seq

def format_line(line):
    words = line.split(" ")
    return int(words[0]),float(words[1]),float(words[2]),float(words[3]),float(words[4])
```

```python
file_path = "./Dataset Test for Yolo Format/datasets/UYZ_2022_OTURUM2_5/detect/labels/train"
#UYZ_2022_OTURUM2_5	2715	1703	481	0	420	0
#C:\Users\Asus\Desktop\Dataset Test for Yolo Format\datasets\UYZ_2022_OTURUM2_5\detect\labels\train
os.listdir(file_path)[0:5]
```

```python


label_file_names_list = os.listdir(file_path)
print(file_path.capitalize())

for label_file_name in label_file_names_list:
    print("İşleniyor",label_file_name)
    
    #çek
    label_file_path = path_sequence(file_path,label_file_name)
    
    f = open(label_file_path,"r")
    lines = f.readlines()
    newlines = []
    f.close()
    
    #değiştir
    for line in lines:
        a,b,c,d,e = format_line(line)
        newtag = targets[a]
        newlines.append(f"{newtag} {b} {c} {d} {e}\n")
        
        if a == targets[a]:
            print(f"AYNI- {a} {targets[a]}")
        else:
            print(f"DEĞİŞTİ- {a} -> {targets[a]}")

    #kaydet
    f = open(label_file_path,"w")
    f.writelines(newlines)
    f.close()
    
```

```python
import cv2
import numpy as np
import os
from tqdm import tqdm
from IPython.display import clear_output
import random
from sklearn.utils import shuffle
```

```python
def path_sequence(root,*directories):
    seq = root
    for directory in directories:
        seq += f"/{directory}"
    return seq

def format_line(line):
    words = line.split(" ")
    return int(words[0]),float(words[1]),float(words[2]),float(words[3]),float(words[4])

def change_line(lbl,x,y,w,h):
    return f"{label_targets[int(lbl)]} {x} {y} {w} {h}\n"
```

```python

dataset_name = "aerialMaritime" ## problem loglarken kullanılıyor değişmeli
root = "./datasets/Aerial Maritime/detect"

labels_path = path_sequence(root,"labels")

for i,c in tqdm( enumerate(["train","test","val"])):
    labels_file = path_sequence(labels_path,c)
    
    label_names = os.listdir(labels_file)
    
    for label_name in label_names:
        
        ## dosyalara ulaş
        print("--",label_name)
         # etiketler
        f = open(path_sequence(labels_file,label_name),"r")
        lines = f.readlines()
        f.close()
        
        newlines = []
        for line in lines:
            label, norm_center_x, norm_center_y, norm_width, norm_height = format_line(line)
            newline = change_line(label,norm_center_x,norm_center_y,norm_width,norm_height)
            newlines.append(newline)
        
        f = open(path_sequence(labels_file,label_name),"w")
        f.writelines(newlines)
        f.close()
```

```python
import xmltodict
```

```python
with open("deneme.xml",encoding='utf-8') as xml_file:
    data_dict = xmltodict.parse(xml_file.read(),encoding='utf-8')
```

```python
data_dict['annotation'].keys()
```

```python
im_width = int(data_dict["annotation"]["size"]["width"])
im_height = int(data_dict["annotation"]["size"]["height"])

for obj in data_dict['annotation']["object"]:
    lbl = obj["name"]
    x1 =float(obj["bndbox"]["xmin"])
    y1 =float(obj["bndbox"]["ymin"]) 
    x2 =float(obj["bndbox"]["xmax"])   
    y2 =float(obj["bndbox"]["ymax"])

    is_inilebilir = obj["attributes"]["attribute"][0]["value"]
    
    
    width = int(x2-x1)
    height = int(y2 - y1)
    center_x = x1 + int(width/2)
    center_y = y1 + int(height/2)
    
    norm_center_x = center_x / im_width
    norm_center_y = center_y / im_height
    norm_width = width / im_width
    norm_height = height / im_height
    
    print(lbl,x1,x2,y1,y2,is_inilebilir)
    print(norm_center_x,norm_center_y,norm_width,norm_height)
    print()
```

#### Tüm xml dosyalarını yolo'ya çevirme işlemi

```python
def path_sequence(root,*directories):
    seq = root
    for directory in directories:
        seq += f"/{directory}"
    return seq
```

```python
from pylabel import importer

xml_files_path = path_sequence("./SpirAl","verisetleri","ornek_videolar","UYZ_2021_etiketli","Annotations")
dataset = importer.ImportVOC(path=xml_files_path)
dataset.export.ExportToYoloV5()
```
